Write a program to find the difference between the sum of all **even elements** and the sum   <br>    of all **odd elements** from the given array, A.

**Note:** The array of elements will be passed as an argument to the function. You don't need to take any input of array.

```python
virat_runs = [67, 54, 12, 34, 77, 89, 101, 120]
```

```python
def even_odd_difference(list_q):
  even_sum = 0
  odd_sum = 0
  for i in list_q:
    if i%2 == 0:
      even_sum += i
    else:
      odd_sum += i
  return even_sum - odd_sum
```

```python
even_odd_difference(virat_runs)
```

```python
even_odd_difference([1,2,3,4,5])
```

```python
sum(list(filter(lambda x: x if x % 2 == 0 else None,virat_runs)))
```

```python
sum(list(filter(lambda x: x if x % 2 != 0 else None,virat_runs)))
```

```python
def even_odd_difference2(list_q):
  return sum(list(filter(lambda x: x if x % 2 == 0 else None,list_q))) - sum(list(filter(lambda x: x if x % 2 != 0 else None,list_q)))
```

```python
even_odd_difference2(virat_runs)
```
