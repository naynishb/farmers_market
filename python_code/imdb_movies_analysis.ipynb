```python
df = movies.merge(directors,left_on = "director_id",right_on = "id",how="left")
```

```python
df.drop(["director_id","Unnamed: 0","id_y"],axis = 1,inplace = True)
```

```python
df.rename({"id_x":"movie_id"},inplace= True,axis = 1)
```

```python
df["revenue"] = (df["revenue"]/1000000).round(2)
df["budget"] = (df["budget"]/1000000).round(2)
```

```python
df.head()
```

```python
df["profit"] = df["revenue"] - df["budget"]
```

```python
df.head()
```

```python
df_profitable_directors = df[df["profit"] > 0]
```

```python
df_profitable_directors.director_name.value_counts().head(10)
```

```python
df_profitable_directors.groupby("director_name")["revenue"].mean()
```

```python
def risky_movie_to_do(director_group):
    director_group["risky"] = director_group["budget"] >= director_group["revenue"].mean()
    return director_group
```

```python
data_risky_directors = df_profitable_directors.groupby("director_name").apply(risky_movie_to_do)
```

```python
data_risky_directors.head(10)
```

```python
data_risky_directors["risky"].sum()
```

```python
risky_movies_list = data_risky_directors[data_risky_directors["risky"]==True]
```
