```python
'''
School grading system, user will input marks
marks more than 90: A
between 75 and 90 : B
between 50 and 74 : C
between 35 and 49 : D
less than 35 : F
'''
```

```python
total_marks = float(input("Enter your marks "))

if total_marks > 90:
  print("A")
elif total_marks >= 75 and total_marks <= 90:
  print("B")
elif total_marks >= 50 and total_marks < 75:
  print("C")
elif total_marks >= 35 and total_marks < 50:
  print("D")
else:
  print("F")
```

```python
'''
Take an integer n as input.
if n is a multiple of only 3 --> print "Fizz"
else if n is a multiple of only 5 --> print "Buzz"
else if n is a multiple of both 3 and 5 --> print "FizzBuzz"
'''
```

```python
N = int(input("Enter a number "))

if N % 3 == 0 and N % 5 == 0:
  print("FizzBuzz")
elif N % 3 == 0:
  print("Fizz")
elif N % 5 == 0:
  print("Buzz")
```
