```python
'''
Question 2 -
1. Take a string as input.
2. Create a dictionary acc to the following criteria -
- There will be one key value pair for each unique character -
- Key will be character and value will be the count of character inside the string

INPUT - "rrsssstttt"
OUTPUT -
{
    "r": 2,
    "s": 4,
    "t": 4
}
'''
```

```python
set("rrsssstttt")
```

```python
"rrsssstttt".count("r")
```

```python
def char_count(string1):
    output = {}
    for char in set(string1):
        output[char] = string1.count(char)
    return output
```

```python
char_count("rrsssstttt")
```

```python
# Create a function that merges two dictionaries. If there are duplicate keys, sum their values.
dict1 = {'a': 1, 'b': 2} # create a copy and use it as it is
dict2 = {'b': 3, 'c': 4} # insert these values in copy
# {'a': 1, 'b': 5, 'c': 4}
```

```python
dict2.items()
```

```python
def merger_dict(dict1,dict2):
  dict1_copy = dict1.copy()
  for key,value in dict2.items():
    if key in dict1_copy:
      dict1_copy[key] += value
    else:
      dict1_copy[key] = value
  return dict1_copy
```

```python
merger_dict(dict1,dict2)
```

```python
'''
Write a function that takes a list of strings and groups them into anagrams.
Return a list of lists, where each inner list contains words that are anagrams of each other.
["eat", "tea", "tan", "ate", "nat", "bat"]

output = [['eat', 'tea', 'ate'], ['tan', 'nat'], ['bat']]
'''
```

```python
def group_anagrams(list_of_words):
  anagrams = {}
  for word in list_of_words:
    key = tuple(sorted(word))
    if key in anagrams:
      anagrams[key].append(word)
    else:
      anagrams[key] = [word]
  return list(anagrams.values())
```

```python
group_anagrams(["eat", "tea", "tan", "ate", "nat", "bat"])
```

```python
def find_duplicates(input_list):
  duplicate_values = []
  unique_values = set(input_list)
  for i in unique_values:
    if input_list.count(i) > 1:
      duplicate_values.append(i)
  return duplicate_values
```

```python
print(find_duplicates([4, 3, 2, 7, 8, 2,2,2,2, 3,3,3,3, 1]))  # Output: [2, 3]
```

```python
def find_duplicates2(input_list):
  count = {}
  duplicate_values = []
  for i in input_list:
    if i in count:
      duplicate_values.append(i)
    else:
      count[i] = 1
  return list(set(duplicate_values ))
```
