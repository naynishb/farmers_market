```python
'''
Question 1 -
Print the following pattern -
*
**
***
****
*****
Input provided will be number of rows.
'''
```

```python
N = int(input("enter number of rows>> "))
for i in range(1,N+1):
  for j in range(i):
    print("*",end = " ")
  print()
```

```python
N = int(input("enter number of rows>> "))
for i in range(1,N+1):
  for j in range(1,i+1):
    print(j,end = " ")
  print()
```

```python
'''
Question 3 -
Print the following pattern -
*
*2
*2*
*2*4
*2*4*
*2*4*6
Input provided will be number of rows.
'''
```

```python
N = int(input("enter number of rows>> "))
for i in range(1,N+1):
  for j in range(1,i+1):
    if j % 2 == 0:
      print(j,end = " ")
    else:
      print("*",end = " ")
  print()
```

```python
'''
Question 4 -
Print the following pattern -
*****
****
***
**
*
Input provided will be number of rows.
'''
```

```python
N = int(input("enter number of rows>> "))
for i in range(1,N+1):
  for j in range((N-i)+1):
    print("*",end = " ")
  print()
```
